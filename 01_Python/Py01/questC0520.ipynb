Q 1-2 회문 분류기

```python
def is_palindrome(word):
    return word == word[::-1]

def palindrome_classifier(word_list):
    result_dict = {"회문O": [], "회문X": []}
    
    for word in word_list:
        if is_palindrome(word):
            result_dict["회문O"].append(word)
        else:
            result_dict["회문X"].append(word)
            
    palindrome_count = len(result_dict["회문O"])
    total_count = len(word_list)
    palindrome_ratio = palindrome_count / total_count if total_count > 0 else 0.0

    return result_dict, palindrome_count, palindrome_ratio

targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']
print(palindrome_classifier(targets))
```

Q 2-2 성적 분석기

```python
def score_analysis(name):
    kor = 88
    eng = 96
    math = 90
    
    c_kor = 91.50
    c_eng = 87.75
    c_math = 87.00
    c_total = 88.75
    
    my_avg = (kor + eng + math) / 3
    
    k_gap = kor - c_kor
    e_gap = eng - c_eng
    m_gap = math - c_math
    t_gap = my_avg - c_total
    
    print("[" + name + " 성적 분석]")
    print(f"국어: {kor}점, 학급 평균({c_kor}점) 대비 {k_gap:+.2f}")
    print(f"영어: {eng}점, 학급 평균({c_eng}점) 대비 {e_gap:+.2f}")
    print(f"수학: {math}점, 학급 평균({c_math}점) 대비 {m_gap:+.2f}")
    print()
    print(f"전과목 평균: {my_avg:.2f}점, 학급 평균({c_total}점) 대비 {t_gap:+.2f}")

score_analysis("설지민")
```
