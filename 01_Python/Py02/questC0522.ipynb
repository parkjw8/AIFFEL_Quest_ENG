```python
def find_min_max(numbers):
    min_value = min(numbers)
    max_value = max(numbers)
    
    def find_min():
        return min_value
        
    def find_max():
        return max_value
        
    return find_min, find_max

numbers = [10, 5, 8, 12, 3, 7]
find_min, find_max = find_min_max(numbers)

print("최솟값:", find_min())  # 출력: 3
print("최댓값:", find_max())  # 출력: 12
```

```python
def counter(func):
    count = 0
    
    def wrapper(*args, **kwargs):
        nonlocal count
        count += 1
        result = func(*args, **kwargs)
        print(f"{func.__name__} 실행횟수: {count}")
        return result
        
    return wrapper

@counter
def say_hello():
    print("Hello Aiffel!")

for i in range(5):
    say_hello()
```
