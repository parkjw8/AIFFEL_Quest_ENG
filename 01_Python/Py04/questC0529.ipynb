```python
import random
```

```python
class Account:
    account_count = 0

    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance
        self.deposit_count = 0

        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)

        self.account_number = f'{num1}-{num2}-{num3}'

        self.deposit_log = []
        self.withdraw_log = []

        Account.account_count += 1

    def get_account_num(self):
        print(f'총 개설된 계좌의 개수 : {Account.account_count}')

    def deposit(self, amount):
        if amount < 1:
            print("입금은 1원 이상만 가능합니다.")
            return

        self.balance += amount
        self.deposit_count += 1
        self.deposit_log.append(f'{amount}원 입금')

        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            self.deposit_log.append(f'이자 {interest}원 지급')

    def withdraw(self, amount):
        if amount < self.balance:
            self.balance -= amount
            self.withdraw_log.append(f'{amount}원 출금')
        else:
            print("잔고 이상으로 출금할 수 없습니다.")

    def display_info(self):
        print(f'은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance}원')

    def deposit_history(self):
        print(f'{self.owner}님의 입금 내역')
        for log in self.deposit_log:
            print(f'- {log}')

    def withdraw_history(self):
        print(f'{self.owner}님의 출금 내역')
        for log in self.withdraw_log:
            print(f'- {log}')

accounts = [Account("A", 1000),
            Account("B", 1100000),
            Account("C", 60000),
            Account("D", 3000000)    ]

for _ in range(5):
    accounts[0].deposit(2000)
accounts[0].deposit_history()
accounts[0].display_info()

for _ in range(5):
    accounts[0].withdraw(2000)
accounts[0].withdraw_history()
accounts[0].display_info()

for account in accounts:
    if account.balance >= 1000000:
        account.display_info()
```

```python
import random
```

```python

class Charactor:
    def __init__(self, name, level, health, attack, defence):
        self.name = name
        self.level = level
        self.health = health
        self.attack = attack
        self.defence = defence

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        dmg = damage - self.defence
        if dmg > 0:
            self.health -= dmg

    def attack_target(self, target):
        dmg = random.randint(1, self.attack)
        self.take_damage(dmg)

class Player(Charactor):
    def __init__(self, name):
        super().__init__(name, level=1, health=100, attack=25, defence=5)
        self.exp = 0

    def gain_experience(self, amount):
        self.exp += amount
        self.level_up()

    def level_up(self):
        while self.exp >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.exp -= 50

class Monster(Charactor):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        atk = random.randint(5, 20) * level
        dfs = random.randint(10, 30) * level
        super().__init__(name, level, health=hp, attack=atk, defence=dfs)

def battle(player, monster):
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if not monster.is_alive():
            break
        monster.attack_target(player)

    if player.is_alive():
        print('전투 승리!')
        player.gain_experience(monster.level * 20)
        return True
    else:
        print('전투 패배')
        return False

def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}
    player_name = input('플레이어의 이름을 입력하세요:')
    player = Player(player_name)
    for name, level in monster_dict.items():
        monster = Monster(name, level)

        victory = battle(player, monster)

        if not victory:
            print('게임오버')
            return

    print('Clear!')

main()
```
