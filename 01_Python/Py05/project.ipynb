```python
from datetime import datetime

def receipt_decorator(func):
    def wrapper(*args, **kwargs):
        print("\n" + "=" * 45)
        print("          [ 모두의 카페 주문 영수증 ]")
        print("=" * 45)
        func(*args, **kwargs)
        print("=" * 45 + "\n")
    return wrapper

class Kiosk:
    def __init__(self):
        self.menu = [
            ['Americano', 2000],
            ['Latte', 2500],
            ['Green Tea', 3000],
            ['Choco Latte', 3500]
        ]
        self.order_menu = []
        self.order_price = []

    def menu_print(self):
        print("\n" + "-" * 15 + " MENU " + "-" * 15)
        for i, item in enumerate(self.menu, start=1):
            name = item[0]
            price = item[1]
            print(f" {i}. {name:<15} | {price:>5}원")
        print(" 0. 주문 종료 및 결제하기")
        print("-" * 36)

    def menu_select(self):
        while True:
            self.menu_print()
            try:
                n = int(input("원하시는 메뉴의 번호를 입력해주세요: "))
            except ValueError:
                print("❌ 숫자로만 입력해주세요.")
                continue

            if n == 0:
                if not self.order_menu:
                    print("🛒 장바구니가 비어있습니다. 프로그램을 종료합니다.")
                    return False
                print("▶ 주문이 완료되었습니다. 결제 단계로 이동합니다.")
                break
            
            if 1 <= n <= len(self.menu):
                while True:
                    temp_select = input("온도를 선택하세요 (1: HOT, 2: ICE): ")
                    if temp_select == '1':
                        temp = "HOT"
                        break
                    elif temp_select == '2':
                        temp = "ICE"
                        break
                    else:
                        print("❌ 잘못된 입력입니다. 1 또는 2를 입력해주세요.")
                
                selected_item = self.menu[n-1]
                full_menu_name = f"{temp} {selected_item[0]}"
                menu_price = selected_item[1]
                
                self.order_menu.append(full_menu_name)
                self.order_price.append(menu_price)
                print(f"✨ [추가 완료] {full_menu_name}(이)가 장바구니에 담겼습니다.")
            else:
                print("❌ 잘못된 메뉴 번호입니다. 다시 선택해주세요.")
        return True

    def pay(self):
        while True:
            print("\n--- 결제 수단 선택 ---")
            print("1. 현금 결제")
            print("2. 카드 결제")
            method = input("결제 수단을 숫자로 선택해주세요 (1 또는 2): ")
            
            if method == '1':
                print("💵 [현금] 결제가 정상 처리되었습니다.")
                break
            elif method == '2':
                print("💳 [카드] 결제가 정상 처리되었습니다.")
                break
            else:
                print("❌ 잘못된 입력입니다. 다시 선택해주세요.")

    @receipt_decorator
    def table(self):
        now = datetime.now()
        time_str = now.strftime("%Y-%m-%d %H:%M:%S")
        print(f" 주문 일시 : {time_str}")
        print("-" * 45)
        print(f" {'상품명':<18} | {'금액':>10}")
        print("-" * 45)
        
        for menu, price in zip(self.order_menu, self.order_price):
            print(f" {menu:<20} | {price:>8}원")
            
        print("-" * 45)
        total_price = sum(self.order_price)
        print(f" {'총 결제 금액':<16} | {total_price:>8}원")


if __name__ == "__main__":
    print("=== [프로젝트 2] 고도화된 키오스크 시스템 실행 ===")
    kiosk = Kiosk()
    
    if kiosk.menu_select():
        kiosk.pay()
        kiosk.table()
```
